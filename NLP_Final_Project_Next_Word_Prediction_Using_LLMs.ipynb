```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import numpy as np
import matplotlib.pyplot as plt

# Check for CUDA availability and set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# List of model names to use
model_names = ["gpt2", "EleutherAI/gpt-neo-125M"]

# Load models and tokenizers
models = []
tokenizers = []
for model_name in model_names:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name).to(device)  # Move model to GPU if available
    models.append(model)
    tokenizers.append(tokenizer)

# Set the maximum length for input (in terms of characters and tokens)
MAX_INPUT_LENGTH = 500
MAX_TOKEN_LENGTH = 512

# Define unwanted tokens to filter out
UNWANTED_TOKENS = ['\n', '<|endoftext|>']

def predict_next_word(model, tokenizer, text, top_k=10, temperature=0.7):
    # Truncate the input text to the maximum length
    if len(text) > MAX_INPUT_LENGTH:
        truncated_text = text[:MAX_INPUT_LENGTH] + "..."  # Truncate and add ellipsis
    else:
        truncated_text = text

    # Encode the input text and limit the tokens
    inputs = tokenizer.encode(text, return_tensors='pt', max_length=MAX_TOKEN_LENGTH, truncation=True).to(device)

    # Create labels for loss calculation
    labels = inputs.clone()
    labels[0, :-1] = -100  # Ignore all but the last token when calculating loss

    with torch.no_grad():
        outputs = model(inputs, labels=labels)

    next_token_logits = outputs.logits[:, -1, :]

    # Adjust logits with temperature
    next_token_logits = next_token_logits / temperature

    # Get the top k tokens
    top_k_logits, top_k_indices = torch.topk(next_token_logits, top_k)

    # Convert logits to probabilities
    probabilities = torch.softmax(top_k_logits, dim=-1).cpu().numpy()

    # Decode top k predictions and filter out unwanted tokens
    top_k_words = []
    filtered_probabilities = []
    for idx, prob in zip(top_k_indices[0].cpu().numpy(), probabilities[0]):
        decoded_word = tokenizer.decode(idx)
        if decoded_word not in UNWANTED_TOKENS:
            top_k_words.append(decoded_word)
            filtered_probabilities.append(prob)

    # Select the most probable next word from the top k predictions
    predicted_word = top_k_words[0] if top_k_words else ""

    # Return truncated text, predicted word, top-k predictions, probabilities, and loss
    loss = outputs.loss.item() if outputs.loss is not None else 0.0
    return truncated_text, predicted_word, top_k_words, filtered_probabilities, loss

datasets = ["pizza.txt", "NLP.txt"]

# Store the losses and perplexities for plotting later
model_datasets_losses = []
model_datasets_perplexities = []
labels = []

# Iterate through each model and dataset
for model_idx, (model, tokenizer) in enumerate(zip(models, tokenizers)):
    model_name = model_names[model_idx]
    print(f"Using model: {model_name}\n")

    for dataset in datasets:
        print(f"Using dataset: {dataset}")

        total_loss = 0
        num_predictions = 0

        # Load the dataset
        with open(dataset, 'r', encoding='latin-1') as file:
            lines = file.readlines()

        for line in lines:
            input_text = line.strip()  # Clean up the line
            if input_text:  # Ensure there's text to process
                truncated_input, predicted_word, top_k_words, filtered_probabilities, loss = predict_next_word(model, tokenizer, input_text)

                total_loss += loss
                num_predictions += 1

                # Print formatted output: input, predicted next word, and loss
                print(f"Input : '{truncated_input}'")
                print(f"Predicted Next Word: '{predicted_word}'")
                print(f"Loss: {loss:.4f}\n")  # Add an extra newline for better readability

                # Visualization of top-k predictions
                if top_k_words:  # Only plot if there are valid top-k words
                    plt.figure(figsize=(8, 4))  # Adjusting plot size
                    plt.bar(top_k_words, filtered_probabilities)
                    plt.xlabel("Predicted Words")
                    plt.ylabel("Probabilities")
                    plt.title(f"Top-{len(top_k_words)} Predictions for '{truncated_input[:20]}...'")
                    plt.xticks(rotation=45)
                    plt.show()

        # Calculate and print average perplexity for the dataset
        if num_predictions > 0:
            average_loss = total_loss / num_predictions
            perplexity = torch.exp(torch.tensor(average_loss)).item()  # Calculate perplexity
            print(f"Average Perplexity for {dataset}: {perplexity:.2f}\n")

            # Store the loss and perplexity for plotting later
            model_datasets_losses.append(average_loss)
            model_datasets_perplexities.append(perplexity)
            labels.append(f"{model_name} - {dataset}")

    # Print a separator between models
    print("=" * 10 + "\n")

# Plot Loss across all models and datasets
plt.figure(figsize=(8, 6))  # Adjusting plot size
plt.bar(labels, model_datasets_losses)
plt.xlabel("Model-Dataset Combination")
plt.ylabel("Loss")
plt.title("Loss Across Models and Datasets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Perplexity across all models and datasets (Vertical bars)
plt.figure(figsize=(8, 6))  # Adjusting plot size
plt.bar(labels, model_datasets_perplexities)
plt.xlabel("Model-Dataset Combination")
plt.ylabel("Perplexity")
plt.title("Perplexity Across Models and Datasets")
plt.xticks(rotation=45)
plt.tight_layout()  # Ensure everything fits without overlap
plt.show()
```

```python
import torch
import numpy as np

def interactive_prediction(models, tokenizers):
    print("\nInteractive Prediction Mode. Type 'exit' to quit.")

    while True:
        # Get user input
        input_text = input("Enter text for next word prediction: ").strip()

        if input_text.lower() == 'exit':
            print("Exiting interactive mode.")
            break

        total_loss = 0  # To store total loss for perplexity calculation
        num_predictions = 0  # To track number of predictions

        for model_idx, (model, tokenizer) in enumerate(zip(models, tokenizers)):
            model_name = model_names[model_idx]
            print(f"\nUsing model: {model_name}")

            # Make prediction using the current model
            truncated_input, predicted_word, top_k_words, filtered_probabilities, loss = predict_next_word(model, tokenizer, input_text)

            total_loss += loss
            num_predictions += 1

            # Display results
            print(f"\nInput Text: '{truncated_input}'")
            print(f"Predicted Next Word: '{predicted_word}'")
            print(f"Loss: {loss:.4f}")

            # Calculate and display perplexity
            if num_predictions > 0:
                avg_loss = total_loss / num_predictions
                perplexity = np.exp(avg_loss)  # Calculate perplexity
                print(f"Perplexity: {perplexity:.2f}\n")

            # Display top-k predictions in a bar chart
            if top_k_words:  # Only plot if there are valid top-k words
                plt.figure(figsize=(8, 4))  # Adjusting plot size
                plt.bar(top_k_words, filtered_probabilities)
                plt.xlabel("Predicted Words")
                plt.ylabel("Probabilities")
                plt.title(f"Top-{len(top_k_words)} Predictions for '{truncated_input[:20]}...'")
                plt.xticks(rotation=45)
                plt.show()

# Call interactive prediction function
interactive_prediction(models, tokenizers)
```
